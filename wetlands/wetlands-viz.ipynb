```python

import ibis
from ibis import _
from cng.utils import *
from cng.h3 import * 
import os
con = ibis.duckdb.connect("duck.db",extensions = ["spatial", "h3"])
os.environ["AWS_ACCESS_KEY_ID"] = ""
os.environ["AWS_SECRET_ACCESS_KEY"] = ""
os.environ["AWS_DEFAULT_REGION"] = "us-west-2"

nwi = con.read_parquet("s3://us-west-2.opendata.source.coop/giswqs/nwi/wetlands/**")
nwi.group_by(_.WETLAND_TYPE).agg(n = _.WETLAND_TYPE.count()).order_by(_.n.desc()).execute()
```
